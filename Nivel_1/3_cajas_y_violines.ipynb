## Grafico de Cajas

- Sirve para visualizar cuartiles





#### Cuando utilizar:
- Cuando quiero comparar diversos grupos a traves de sus estadisticos (media, mediana y cuartiles)
- Detectar datos aberrantes dentro de alguna categoria

#### Cuando no utilizar:
- Cuando quiero ver la forma exacta de mi destribuicion.
- Cuando quiero ver datos puntuales de mi dataset

#### Consejos:
- Es indiferencte hacer cajas verticales u horizontales. Depende de cuantas quiero mostrar y el tamaño de las etiquetas.

#### Ejemplos de Uso
- Si soy dueño de una página web, me gustaría saber cual es la diferencia del tiempo en que las personas visitan mi página desde un Celular, Tablet o PC de escritorio.
- En un banco, me interesaría saber cuanto dinero almacenan diferentes tipos de cuentas bancarias

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/titanic.csv')
df
```

```python
# Se eliminan los datos nulos de la columna 'Age'
df = df.dropna(subset=['Age'], axis='index')
df
```

```python
# Grafico de cajas en pandas
df.boxplot(
    column='Age',
    by='Sex'
)
```

```python
# Grafico de cajas en Seaborn
sns.boxplot(
    df,
    x='Sex',
    y='Age'
)
```

## Grafico de violines

Los gráficos de cajas no nos dicen nada acerca de la forma de la distribución de los datos. Solamente de los estadísticos de la distribución. Esto hace que puedan ser engañosos (ver imagen). Por esto un excelente acompañante o reemplazo de los gráficos de cajas, es el gráfico de violines. Nos permite ver cómo están distribuidos los datos, manteninedo todas las ventajas de un diagrama de cajas. La única desventaja es que puese ser más complejo de entender para una persona inexperta.



**Eje X**: Variable categórica

**Eje Y**: Variable numérica

#### Cuando utilizar
- Quiero comparar grupos a través de sus estadísticos (media, mediana y percentiles) y sus distribuciones al mismo tiempo

#### Cuando no utilizar
- Si mis puntos de interés son los puntos individuales de mi dataset
- Si no me interesa la forma de la distribución de mis datos, prefiero un diagrama de cajas convencional.

#### Referencias
- https://datavizcatalogue.com/ES/metodos/diagrama_de_violin.html
- https://blog.bioturing.com/2018/05/16/5-reasons-you-should-use-a-violin-graph/

```python
# Violin plot con Seaborn
sns.violinplot(
    df,
    x='Sex',
    y='Age'
)
```

```python
sns.violinplot(
    df,
    x='Sex',
    y='Age',
    inner='quartiles' # quartiles, stick, points
)
```
