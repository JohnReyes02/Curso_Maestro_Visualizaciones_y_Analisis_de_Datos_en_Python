## Grafico de Cajas

- Sirve para visualizar cuartiles





#### Cuando utilizar:
- Cuando quiero comparar diversos grupos a traves de sus estadisticos (media, mediana y cuartiles)
- Detectar datos aberrantes dentro de alguna categoria

#### Cuando no utilizar:
- Cuando quiero ver la forma exacta de mi destribuicion.
- Cuando quiero ver datos puntuales de mi dataset

#### Consejos:
- Es indiferencte hacer cajas verticales u horizontales. Depende de cuantas quiero mostrar y el tamao de las etiquetas.

#### Ejemplos de Uso
- Si soy dueo de una pgina web, me gustara saber cual es la diferencia del tiempo en que las personas visitan mi pgina desde un Celular, Tablet o PC de escritorio.
- En un banco, me interesara saber cuanto dinero almacenan diferentes tipos de cuentas bancarias

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/titanic.csv')
df
```

```python
# Se eliminan los datos nulos de la columna 'Age'
df = df.dropna(subset=['Age'], axis='index')
df
```

```python
# Grafico de cajas en pandas
df.boxplot(
    column='Age',
    by='Sex'
)
```

```python
# Grafico de cajas en Seaborn
sns.boxplot(
    df,
    x='Sex',
    y='Age'
)
```

## Grafico de violines

Los grficos de cajas no nos dicen nada acerca de la forma de la distribucin de los datos. Solamente de los estadsticos de la distribucin. Esto hace que puedan ser engaosos (ver imagen). Por esto un excelente acompaante o reemplazo de los grficos de cajas, es el grfico de violines. Nos permite ver cmo estn distribuidos los datos, manteninedo todas las ventajas de un diagrama de cajas. La nica desventaja es que puese ser ms complejo de entender para una persona inexperta.



**Eje X**: Variable categrica

**Eje Y**: Variable numrica

#### Cuando utilizar
- Quiero comparar grupos a travs de sus estadsticos (media, mediana y percentiles) y sus distribuciones al mismo tiempo

#### Cuando no utilizar
- Si mis puntos de inters son los puntos individuales de mi dataset
- Si no me interesa la forma de la distribucin de mis datos, prefiero un diagrama de cajas convencional.

#### Referencias
- https://datavizcatalogue.com/ES/metodos/diagrama_de_violin.html
- https://blog.bioturing.com/2018/05/16/5-reasons-you-should-use-a-violin-graph/

```python
# Violin plot con Seaborn
sns.violinplot(
    df,
    x='Sex',
    y='Age'
)
```

```python
sns.violinplot(
    df,
    x='Sex',
    y='Age',
    inner='quartiles' # quartiles, stick, points
)
```
