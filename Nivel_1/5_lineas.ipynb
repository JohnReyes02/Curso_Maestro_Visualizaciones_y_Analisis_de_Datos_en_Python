## Gráfico de Línea
Gráfico que representa valores numéricos conectados por líneas de izquierda a derecha. Tienen un objetivo similar al gráfico de barras. Nos sirve para mostrar diferencias de variables numéricas a través de mi variable en el Eje X. Sin embargo, es más útil cuando queremos mostrar tendencias de subida o de bajada de mi variable numérica. A diferencia del gráfico de barras, el Eje X de izquierda a derecha suele representar una unidad de progresión ordenada de menor a mayor (ejemplo: días, meses,años, horas; es decir usualmente unidades de tiempo). Es decir, variables categóricas ordinales. A diferencia del gráfico de barras, yo puedo tener muchas categorías o puntos en el Eje X ya que no necesito dibujar barras en cada punto. El gráfico puede tener varias líneas que representan diferentes grupos. En este caso se denomina gráfico miltilínea. Y obviamente, cada línea tiene que estar correctamente diferenciada.

**Eje X**: Variable categórica ordinal.

**Eje Y**: Variable numérica.

**Otros nombres**: Cuando mi cantidad de valores en el Eje X es muy alta y es una variable temporal, se puede denominar Serie de Tiempo.



#### Cuando Utilizar
- Para analizar tendencias de subida o de bajada en mis variables.
- Para comparar diferencias de varibles numéricas entre uno o más grupos en mis datos.
- Para cuando quiero visualizar cambios de magnitud a lo largo de mi Eje X. Sin embargo, si tengo pocos puntos, un gráfico de barras es mejor alternativa si este es mi único objetivo.

#### Cuando no utilizar
- Si mi varible en el Eje X no tiene una progresión. En ese caso, tendría que pensar en un Gráfico de Barras.
- Si tengo menos de tres puntos para graficar.
- Si quiero comparar proporciones.
- Si quiero obtener información sobre distribuciones.

#### Ejemplos de Uso
- Si soy dueño de una página web, me gustaría saber como ha cambiado el número de visitas entre diferentes años.
- En un banco, me interesaría saber cuántas transacciones ocurren mensualmente.

#### Consejos
- No es necesario que el Eje Y empiece en el valor del "cero" (a diferencia del gráfico de barras). He ahí la razón por la que no es tan útil como el diagrama de barras para comparar magnitudes.
- No interpoles o suavices la curva. Esto podría dar una falsa ilusión de valores inexistentes. Siempre une cada punto con una línea recta.
- Si no hay una gran cantidad de puntos, añade una clave visual para identificar correctamente a que parte de la línea pertenece cada punto (un triángulo, un cuadrado, un círculo, una equis, etc).
- No utilices colores innecesariamente.
- Si vas a hacer un gráfico miltilínea para codificar una variable categórica adicional:
    - Procura que no existan muchas líneas.
    - Procura que las variables que cada línea representan sean diferenciables (por ejemplo, a través de colores diferentes, o a través de una forma diferente en los puntos).

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/tienda_ventas.csv')
df
```

```python
# Dejamos  la primer columna como indice
df = pd.read_csv('../data/tienda_ventas.csv', index_col='id')
df
```

```python
# Agregamos la columna mes
df['mes'] = df['date'].apply(lambda x: x[:7])
df
```

```python
# Ventas por mes en todas las tiendas tomando en cuenta todos los productos
por_mes = df[['mes', 'sales']].groupby('mes').sum()
por_mes
# Otra forma
#por_mes = df.groupby('mes').agg({
#    'sales': 'sum'
#})
#por_mes
```

```python
por_mes.plot.line(
    y='sales'
)
```

```python
# Colocando marcadores
por_mes.plot.line(
    y='sales',
    style='.-',
    figsize=(10, 5)
)
```

```python
# Colocando línea punteada
por_mes.plot.line(
    y='sales',
    style='.-',
    linestyle='dotted'
)
```

```python
import plotly.express as px
```

```python
fig = px.line(
  por_mes,
  x=por_mes.index.values,
  y='sales'
)
fig.show()
```

## Gráfico Multilínea

#### Objetivo actual
- En el 2015, cómo se compara la venta de 3 familias de productos diferentes (MEATS, SEAFOOD y DIARY)


```python
df
```

```python
df['anio'] = df['date'].apply(lambda x: x[:4])
df
```

```python
df_2015 = df[
  df['anio'] == '2015'
]
df_2015
```

```python
df_2015 = df[
  (df['anio'] == '2015') &
  (df['family'].isin(['SEAFOOD', 'MEATS', 'DAIRY']))
]
df_2015
```

```python
# Agrupo las ventas por mes y por familia
agrupacion = df_2015[['mes', 'family', 'sales']].groupby(['mes', 'family']).sum()
agrupacion

# Otra forma
#agrupacion = df_2015.groupby(['mes', 'family']).agg({
#  'sales': 'sum'
#})
#agrupacion
```

```python
# Pivoteo la tabla para que mis categorias pasen a ser mis columnas
pivotado = agrupacion.unstack()
pivotado

# Otra forma
# Se realiza un pivotado para dejar cada familia como columna
#pivotado = agrupacion.reset_index(1)
#pivotado
```

```python
# Elimino un nivel en las columnas, no lo necesito
pivotado.columns = pivotado.columns.droplevel()
pivotado

#pivotado
#pivotado = agrupacion.reset_index(1).pivot(columns='family')

#pivotado
```

```python
# Gráfico de líneas múltiples categorías
pivotado.plot.line(
  style='.-'
)
```

```python
# Gráfico de líneas con múltiples categorías con subplots
pivotado.plot.line(
  style='.-',
  subplots=True
)
```

## Gráfico de Áreas
Combina la utilidad del gráfico de líneas con la del gráfico de barras. Me permite comparar diferencias entre magnitudes y hallar tendencias a lo largo de una progresión continua. Es por esto que en un gráfico de Areas, el Eje Y si tiene que empezar en el "cero" (como el gráfico de baras). Básicamente es un gráfico de líneas con el area bajo la curva de un color sólido.

**Eje X**: Variable categórica.

**Eje Y**: Variable numérica.

**Otros nombres**: Area Chart

**Subtipos**:
1. Stacked Area Chart (Gráfico de Áreas Apilados): Múltiples líneas que acumulan sus valores numéricos del Eje Y.
2. Overlapped Area Chart (Gráfico de Áreas Agrupados): Múltiples líneas que no acumulan sus valores numéricos del Eje Y.



#### Cuando utilizar
- Cuando quiero comparar magnitudes y observar tendencias a lo largo de mi Eje X.

#### Referencias
- https://datavizcatalogue.com/ES/metodos/grafica_de_area.html
- https://chartio.com/learn/charts/area-chart-complete-guide/

```python
# Grafico de area comun
por_mes.plot.area(
    y='sales'
)
```

#### Grafico de Areas apiladas (Stacked Area Chart)
- Apila areas de diferentes grupos una encima de la otra, SUMANDO sus magnitudes

**TRADEOFF**
- Pierdo precision al momento de intentar saber la magnitud exacta de la variable en el Eje Y de mis grupos individuales.
- Gano poder comparar proporcionalmente las diferentes areas que representan los grupos.

```python
pivotado.plot.area(
    stacked=True
)
```

#### Grafico de areas agrupadas
Es un gráfico multilínea que pinta el área bajo la curva de cada línea. No suma las magnitudes. Simplemente pinta debajo de las líneas.

```python
pivotado.plot.area(
    stacked=False,
    style='.-'
)
```
