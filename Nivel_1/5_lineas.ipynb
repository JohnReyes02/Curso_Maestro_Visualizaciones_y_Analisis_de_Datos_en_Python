## Grfico de Lnea
Grfico que representa valores numricos conectados por lneas de izquierda a derecha. Tienen un objetivo similar al grfico de barras. Nos sirve para mostrar diferencias de variables numricas a travs de mi variable en el Eje X. Sin embargo, es ms til cuando queremos mostrar tendencias de subida o de bajada de mi variable numrica. A diferencia del grfico de barras, el Eje X de izquierda a derecha suele representar una unidad de progresin ordenada de menor a mayor (ejemplo: das, meses,aos, horas; es decir usualmente unidades de tiempo). Es decir, variables categricas ordinales. A diferencia del grfico de barras, yo puedo tener muchas categoras o puntos en el Eje X ya que no necesito dibujar barras en cada punto. El grfico puede tener varias lneas que representan diferentes grupos. En este caso se denomina grfico miltilnea. Y obviamente, cada lnea tiene que estar correctamente diferenciada.

**Eje X**: Variable categrica ordinal.

**Eje Y**: Variable numrica.

**Otros nombres**: Cuando mi cantidad de valores en el Eje X es muy alta y es una variable temporal, se puede denominar Serie de Tiempo.



#### Cuando Utilizar
- Para analizar tendencias de subida o de bajada en mis variables.
- Para comparar diferencias de varibles numricas entre uno o ms grupos en mis datos.
- Para cuando quiero visualizar cambios de magnitud a lo largo de mi Eje X. Sin embargo, si tengo pocos puntos, un grfico de barras es mejor alternativa si este es mi nico objetivo.

#### Cuando no utilizar
- Si mi varible en el Eje X no tiene una progresin. En ese caso, tendra que pensar en un Grfico de Barras.
- Si tengo menos de tres puntos para graficar.
- Si quiero comparar proporciones.
- Si quiero obtener informacin sobre distribuciones.

#### Ejemplos de Uso
- Si soy dueo de una pgina web, me gustara saber como ha cambiado el nmero de visitas entre diferentes aos.
- En un banco, me interesara saber cuntas transacciones ocurren mensualmente.

#### Consejos
- No es necesario que el Eje Y empiece en el valor del "cero" (a diferencia del grfico de barras). He ah la razn por la que no es tan til como el diagrama de barras para comparar magnitudes.
- No interpoles o suavices la curva. Esto podra dar una falsa ilusin de valores inexistentes. Siempre une cada punto con una lnea recta.
- Si no hay una gran cantidad de puntos, aade una clave visual para identificar correctamente a que parte de la lnea pertenece cada punto (un tringulo, un cuadrado, un crculo, una equis, etc).
- No utilices colores innecesariamente.
- Si vas a hacer un grfico miltilnea para codificar una variable categrica adicional:
    - Procura que no existan muchas lneas.
    - Procura que las variables que cada lnea representan sean diferenciables (por ejemplo, a travs de colores diferentes, o a travs de una forma diferente en los puntos).

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/tienda_ventas.csv')
df
```

```python
# Dejamos  la primer columna como indice
df = pd.read_csv('../data/tienda_ventas.csv', index_col='id')
df
```

```python
# Agregamos la columna mes
df['mes'] = df['date'].apply(lambda x: x[:7])
df
```

```python
# Ventas por mes en todas las tiendas tomando en cuenta todos los productos
por_mes = df[['mes', 'sales']].groupby('mes').sum()
por_mes
# Otra forma
#por_mes = df.groupby('mes').agg({
#    'sales': 'sum'
#})
#por_mes
```

```python
por_mes.plot.line(
    y='sales'
)
```

```python
# Colocando marcadores
por_mes.plot.line(
    y='sales',
    style='.-',
    figsize=(10, 5)
)
```

```python
# Colocando lnea punteada
por_mes.plot.line(
    y='sales',
    style='.-',
    linestyle='dotted'
)
```

```python
import plotly.express as px
```

```python
fig = px.line(
  por_mes,
  x=por_mes.index.values,
  y='sales'
)
fig.show()
```

## Grfico Multilnea

#### Objetivo actual
- En el 2015, cmo se compara la venta de 3 familias de productos diferentes (MEATS, SEAFOOD y DIARY)


```python
df
```

```python
df['anio'] = df['date'].apply(lambda x: x[:4])
df
```

```python
df_2015 = df[
  df['anio'] == '2015'
]
df_2015
```

```python
df_2015 = df[
  (df['anio'] == '2015') &
  (df['family'].isin(['SEAFOOD', 'MEATS', 'DAIRY']))
]
df_2015
```

```python
# Agrupo las ventas por mes y por familia
agrupacion = df_2015[['mes', 'family', 'sales']].groupby(['mes', 'family']).sum()
agrupacion

# Otra forma
#agrupacion = df_2015.groupby(['mes', 'family']).agg({
#  'sales': 'sum'
#})
#agrupacion
```

```python
# Pivoteo la tabla para que mis categorias pasen a ser mis columnas
pivotado = agrupacion.unstack()
pivotado

# Otra forma
# Se realiza un pivotado para dejar cada familia como columna
#pivotado = agrupacion.reset_index(1)
#pivotado
```

```python
# Elimino un nivel en las columnas, no lo necesito
pivotado.columns = pivotado.columns.droplevel()
pivotado

#pivotado
#pivotado = agrupacion.reset_index(1).pivot(columns='family')

#pivotado
```

```python
# Grfico de lneas mltiples categoras
pivotado.plot.line(
  style='.-'
)
```

```python
# Grfico de lneas con mltiples categoras con subplots
pivotado.plot.line(
  style='.-',
  subplots=True
)
```

## Grfico de reas
Combina la utilidad del grfico de lneas con la del grfico de barras. Me permite comparar diferencias entre magnitudes y hallar tendencias a lo largo de una progresin continua. Es por esto que en un grfico de Areas, el Eje Y si tiene que empezar en el "cero" (como el grfico de baras). Bsicamente es un grfico de lneas con el area bajo la curva de un color slido.

**Eje X**: Variable categrica.

**Eje Y**: Variable numrica.

**Otros nombres**: Area Chart

**Subtipos**:
1. Stacked Area Chart (Grfico de reas Apilados): Mltiples lneas que acumulan sus valores numricos del Eje Y.
2. Overlapped Area Chart (Grfico de reas Agrupados): Mltiples lneas que no acumulan sus valores numricos del Eje Y.



#### Cuando utilizar
- Cuando quiero comparar magnitudes y observar tendencias a lo largo de mi Eje X.

#### Referencias
- https://datavizcatalogue.com/ES/metodos/grafica_de_area.html
- https://chartio.com/learn/charts/area-chart-complete-guide/

```python
# Grafico de area comun
por_mes.plot.area(
    y='sales'
)
```

#### Grafico de Areas apiladas (Stacked Area Chart)
- Apila areas de diferentes grupos una encima de la otra, SUMANDO sus magnitudes

**TRADEOFF**
- Pierdo precision al momento de intentar saber la magnitud exacta de la variable en el Eje Y de mis grupos individuales.
- Gano poder comparar proporcionalmente las diferentes areas que representan los grupos.

```python
pivotado.plot.area(
    stacked=True
)
```

#### Grafico de areas agrupadas
Es un grfico multilnea que pinta el rea bajo la curva de cada lnea. No suma las magnitudes. Simplemente pinta debajo de las lneas.

```python
pivotado.plot.area(
    stacked=False,
    style='.-'
)
```
