## Heatmap (Mapa de Calor)
Es un grfico X vs Y con dos usos diferentes dependiendo del tipo de las variables en los ejes.

**Si las dos variables son numricas:** El heatmap es una matriz que representa como se distribuyen las dos variables vindolas como un par de puntos. Es decir, que par de valores (x,y) tienen una mayor o menor frecuencia. Lo podemos ver como un Scatter Plot agrupado en cuadrculas equidistantes. En donde una cuadrcula con una tonalidad ms oscura me indica que existen ms de esos pares de puntos (x,y) dentro de esa cuadrcula. Y una cuadrcula con una tonalidad ms clara que indica que existen menos pares de puntos (x,y) dentro de esa cuadrcula. Por lo general, se lo utiliza cuando tengo un Scatter Plot con demasiados puntos o cuando quiero tener informacin sobre la distribucin de las variables como un par. Y no individualmente.
**Otros nombres**: Si las cuadrculas son hexagonales, se le denomina Hexabin. Y la funcionalidad es la misma.



**Si las dos variables son categricas y codifican una variable numrica**: El Heatmap es una matriz con tantas celdas como valores distintos categricos tenga en X y en Y. Una celda con una tonalidad ms oscura me indica que el valor de ese punto X vs Y es mayor. **Ejemplos de Uso**: Como dueo de un supermercado me gustara saber qu tipo de productos se han vendido ms y en cuales supermercados. En este caso mi eje X seran los tipos de productos, mi eje Y seran los nombres de los supermercados, y la celda con su intensidad de color codificaran el valor de ventas de ese produto en ese supermercado.



**Otros Nombres**:
Existe algo denominado Heatmap de Correlaciones o Matriz de Correlacin. En donde ambos ejes tienen nombres de variables (o columnas). Y la celda muestra el coeficiente de correlacin entre esas dos variables.



Otro ejemplo de un Heatmap seria el grafico del nmero de contribuciones de un usuario de Github. Rpidamente podemos ver que dias hemos hecho mas contribuciones. Es un grfico muy intuitivo y til para mostrar una cantidad grande de informacin de forma efectiva.



#### Cundo Utilizar

- Cuando quiero observar la distribucin de dos variables como un par de puntos.
- Cuando quiero visualizar de forma intuitiva mucha informacin de una gran cantidad de categoras en un solo grfico.

#### Consejos
- Usualmente la tonalidad ms oscura representar un mayor valor.
- Es recomendable aadir una leyenda que muestre el significado de las tonalidades. Los rangos de valores que cada tonalidad codifica.
- Cuando utilizamos variables categricas es comm mostrar el valor que codifica la celda dentro de la celda.
- Es una alternativa cuando quiero hacer un grfico de barras mltiples, y tengo muchas categoras. Como el ejemplo que mencionamos. Si tengo muchos supermercados, o muchos tipos de productos diferentes. Un grfico de barras ser demasiado denso.
- Si tengo muy pocos puntos en mi dataset, es preferible un scatter plot. O si tengo solo una o dos categoras es mejor un diagrama de barras.

```python
 import pandas as pd
```

```python
df = pd.read_csv('../data/pinguinos.csv')
df = df.dropna()
df
```

```python
import seaborn as sns
```

```python
# Heatmap con jointplot

sns.jointplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    kind='hist',
    joint_kws={
        'bins':20
    },
    marginal_kws={
        'bins': 20,
        'fill': False
    },
    # cbar=True, # Muestra la leyenda de la barra de colores
)
```

```python
sns.jointplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    kind='hex',
    joint_kws={
        'bins':20
    },
    marginal_kws={
        'bins': 20,
        'fill': False
    },
    # cbar=True, # Muestra la leyenda de la barra de colores
    gridsize=30
)
```

```python
# Otra forma de hace un heatmap

df.plot.hexbin(
    x='flipper_length_mm',
    y='culmen_length_mm',
    gridsize=30
)
```

## Heatmap Categrico

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/tienda_ventas.csv')
df = df.dropna()
df
```

```python
# Creamos una nueva columna con el ao y el mes
df['periodo'] = df['date'].apply(lambda x: x[:7])
df
```

```python
# Filtramos las tiendas 4, 44, 45, 46, 47, 48, 49, 50
df2 = df[df['store_nbr'].isin([4, 44, 45, 46, 47, 48, 49, 50])]
df2
```

```python
# Agrupamos por las ventas por periodos de cada tienda.
df2_agrupado = df2.groupby(['periodo', 'store_nbr']).agg({
    'sales': 'sum'
})
df2_agrupado
```

```python
df2_agrupado = df2_agrupado.unstack()
df2_agrupado
```

```python
matriz = df2_agrupado.transpose()
matriz
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(
    figsize=(15, 8)
)

sns.heatmap( # Solo para variables categricas
    matriz,
    cmap='Blues', # Cambia a escalation de azules
    square=True,
    linewidth=0.1,
    # annot=True para poner el valor dentro de cada casilla, no se pone aqu porque los valores son grandes
)
```
