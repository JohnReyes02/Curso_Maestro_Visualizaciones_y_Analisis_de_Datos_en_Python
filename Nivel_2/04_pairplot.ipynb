## Pair Plot
Es otro grfico compuesto. Es un grfico de "todos contra todos". Compara al mismo tiempo todas las variables numricas de mi dataset a travs de diversos Scatter plots arreglados en un grid (N x N). Siendo N el nmero de variables en mi dataset. En cada fila del grid, se compara en diversos scatter plot, una variable numrica VERSUS todas las dems. La diagonal del grid va a tener un histograma o un KDE de esa varible. (Ya que no tendra utilidad hacer un scatter plot donde ambos ejes correspondan a la misma variable)



#### Cuando Utilizar
- Quiero analizar si existe algn tipo de correlacin entre alguna de todas mis variables numricas.
- Quiero observar la distribucin de todas mis variables numricas al mismo tiempo.
- Es uno de los grficos ms tilies al hacer anlisis exploratorio de datos ya que me muestra un resumen de todos mis datos numricos.

#### Consejos
- Si tengo muchas varibles de inters, el GRID se volver demasiado grande. En estos casos puedo escoger un subconjunto de mis columnas para formar el Pair Plot.
- Puedo introducir una tercera variable al grfico con las siguientes tcnicas:
    - Diferentes colores o formas en los "puntos" para introducir una variable categrica.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
```

```python
# leemos los datos con pandas y eliminamos las filas NaNs 
df = pd.read_csv('../data/pinguinos.csv')
df = df.dropna()
df
```

```python
sns.pairplot(
    data=df
)
```

```python
sns.pairplot(
    data=df,
    hue='species'
)
```

```python
sns.pairplot(
    data=df,
    hue='species',
    kind='kde'
)
```

```python
sns.pairplot(
    data=df,
    hue='species',
    kind='hist'
)
```
