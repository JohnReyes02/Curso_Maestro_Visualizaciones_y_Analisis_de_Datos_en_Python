## Pair Plot
Es otro gráfico compuesto. Es un gráfico de "todos contra todos". Compara al mismo tiempo todas las variables numéricas de mi dataset a través de diversos Scatter plots arreglados en un grid (N x N). Siendo N el número de variables en mi dataset. En cada fila del grid, se compara en diversos scatter plot, una variable numérica VERSUS todas las demás. La diagonal del grid va a tener un histograma o un KDE de esa varible. (Ya que no tendría utilidad hacer un scatter plot donde ambos ejes correspondan a la misma variable)



#### Cuando Utilizar
- Quiero analizar si existe algún tipo de correlación entre alguna de todas mis variables numéricas.
- Quiero observar la distribución de todas mis variables numéricas al mismo tiempo.
- Es uno de los gráficos más útilies al hacer análisis exploratorio de datos ya que me muestra un resumen de todos mis datos numéricos.

#### Consejos
- Si tengo muchas varibles de interés, el GRID se volverá demasiado grande. En estos casos puedo escoger un subconjunto de mis columnas para formar el Pair Plot.
- Puedo introducir una tercera variable al gráfico con las siguientes técnicas:
    - Diferentes colores o formas en los "puntos" para introducir una variable categórica.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
```

```python
# leemos los datos con pandas y eliminamos las filas NaNs 
df = pd.read_csv('../data/pinguinos.csv')
df = df.dropna()
df
```

```python
sns.pairplot(
    data=df
)
```

```python
sns.pairplot(
    data=df,
    hue='species'
)
```

```python
sns.pairplot(
    data=df,
    hue='species',
    kind='kde'
)
```

```python
sns.pairplot(
    data=df,
    hue='species',
    kind='hist'
)
```
