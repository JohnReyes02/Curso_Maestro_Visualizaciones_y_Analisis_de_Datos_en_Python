## Joinplot
Es un grfico compuesto por dos tipos de grficos diferentes. Es un Scatter Plot que al mismo tiempo muestra histogramas (o KDE) con la distribucin de cada variable. Para este grfico, ambas variables deben ser numricas.

**Eje X**: Variable numrica.

**Eje Y**: Variable numrica.



### En esta clase
#### Datos
Informacin sobre especies de pinguinos

#### Objetivos 
1. Encontrar alguna relacin entre las caractersticas de los pinguinos y al mismo tiempo ver la distribucin de las variables

### Referencias
- https://github.com/lkuffo/data-viz/blob/main/NIVEL2/3_jointplot.ipynb
- https://seaborn.pydata.org/generated/seaborn.jointplot.html

```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
```

```python
df = pd.read_csv('../data/pinguinos.csv')
df = df.dropna()
df
```

```python
sns.jointplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    hue='species' # Se codifica una variable categorica de diferente color
)
```

```python
# grfico de densidades con la regresin lineal
sns.jointplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    kind='reg'
    # hue='species' # Se codifica una variable categorica de diferente color
)
```
