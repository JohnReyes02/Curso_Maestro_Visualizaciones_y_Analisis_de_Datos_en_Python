# Gráfico de Regresión
Es un Scatter Plot con una línea de tendencia (Line of Best fit). Esta línea representa matemáticamente una regresión lineal con la forma *y = mx + b*. En otras palabras, es una recta con la que yo puedo "predecir" qué valor de "y" tendría un nuevo valore de "x". Para este gráfico, ambas variables deben ser numéricas.

**Eje X**: Variable Numérica.
**Eje Y**: Variable Numérica.



**Ejemplo de Uso**:
- Como doctor, me gustaría saber si el nivel de colesterol de una persona se correlaciona con su peso.


#### Cuando Utilizar
- Cuando quiero comprobar si existe algún tipo de correlación entre dos variables.
- Cuando quiero observar la distribución de dos variables al mismo tiempo.

#### Consejos
- No es necesario que el eje Y empiece en el valor del "cero".
- No utilices colores innecesariamente. Se pueden pintar los puntos para introducir una tercera variable categórica al gráfico.
- Puedo introducir una tercera o cuarta variable al gráfico con las siguientes técnicas:
  - Difrentes formas en los "puntos" para introducir una variable categórica.
  - Convertirlo en una gráfica de burbujas para introducir una variable numérica.
  - Jugar con la intensidad del color de cada punto para introducir una variable numérica.
- Si tengo muy pocos puntos en mi dataset, podemos coger una muestra, o jugar con la opacidad de los puntos.

```python
import pandas as pd
import seaborn as sns # el autor del curso manifiesta que esta libreria es mas util cuando va representar la regresion

sns.set_style('whitegrid')
```

```python
df = pd.read_csv('../data/pinguinos.csv')
df = df.dropna()
# df = df[df['species'] == 'Adelie']
df
```

```python
%pip install scipy
```

```python
# Calculamos el coeficiente de correlacon
import scipy as sp
import matplotlib.pyplot as plt
```

```python
r, p = sp.stats.pearsonr(df['flipper_length_mm'], df['culmen_length_mm']) # Calculo del Coeficiente de correlacion lineal
r, p
```

**El valor P**

El valor P valida que mi coeficiente tenga significancia estadistica. Si mi valor de P es mayor que 0.05, mi coeficiente queda invalidado. O sencillamente, no existe una correlacion.

```python
sns.regplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    scatter_kws={
        's': 10
    }
)

texto = 'r=' + str(r)[:5]

plt.text(225, 42, texto, color='black')
plt.title('Longitud de las Aletas vs Longitud del Pico en Pinguinos')
```
