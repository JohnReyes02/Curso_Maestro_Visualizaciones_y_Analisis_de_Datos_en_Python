# Grfico de Regresin
Es un Scatter Plot con una lnea de tendencia (Line of Best fit). Esta lnea representa matemticamente una regresin lineal con la forma *y = mx + b*. En otras palabras, es una recta con la que yo puedo "predecir" qu valor de "y" tendra un nuevo valore de "x". Para este grfico, ambas variables deben ser numricas.

**Eje X**: Variable Numrica.
**Eje Y**: Variable Numrica.



**Ejemplo de Uso**:
- Como doctor, me gustara saber si el nivel de colesterol de una persona se correlaciona con su peso.


#### Cuando Utilizar
- Cuando quiero comprobar si existe algn tipo de correlacin entre dos variables.
- Cuando quiero observar la distribucin de dos variables al mismo tiempo.

#### Consejos
- No es necesario que el eje Y empiece en el valor del "cero".
- No utilices colores innecesariamente. Se pueden pintar los puntos para introducir una tercera variable categrica al grfico.
- Puedo introducir una tercera o cuarta variable al grfico con las siguientes tcnicas:
  - Difrentes formas en los "puntos" para introducir una variable categrica.
  - Convertirlo en una grfica de burbujas para introducir una variable numrica.
  - Jugar con la intensidad del color de cada punto para introducir una variable numrica.
- Si tengo muy pocos puntos en mi dataset, podemos coger una muestra, o jugar con la opacidad de los puntos.

```python
import pandas as pd
import seaborn as sns # el autor del curso manifiesta que esta libreria es mas util cuando va representar la regresion

sns.set_style('whitegrid')
```

```python
df = pd.read_csv('../data/pinguinos.csv')
df = df.dropna()
# df = df[df['species'] == 'Adelie']
df
```

```python
%pip install scipy
```

```python
# Calculamos el coeficiente de correlacon
import scipy as sp
import matplotlib.pyplot as plt
```

```python
r, p = sp.stats.pearsonr(df['flipper_length_mm'], df['culmen_length_mm']) # Calculo del Coeficiente de correlacion lineal
r, p
```

**El valor P**

El valor P valida que mi coeficiente tenga significancia estadistica. Si mi valor de P es mayor que 0.05, mi coeficiente queda invalidado. O sencillamente, no existe una correlacion.

```python
sns.regplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    scatter_kws={
        's': 10
    }
)

texto = 'r=' + str(r)[:5]

plt.text(225, 42, texto, color='black')
plt.title('Longitud de las Aletas vs Longitud del Pico en Pinguinos')
```
