## Grfico de Dispersin (Scatter Plot)
Grfico que muestra los valores de dos variables de un mismo registro en nuestro dataset al mismo tiempo. En cada Eje del grfico se coloca una variable. Se dibuja en el canvas un punto, en la coordenada correspondiente al valor de las variables por cada observacin de nuestra dataset. Su objetivo es detectar si existe o no alguna correlacin entre las variables en los ejes. Estos ejes usualmente son variables numricas pero tambin se pueden utilizar variables categricas ordinales.

**Eje X**: Variable numrica o categrica 

**Eje Y**: Variable numrica o categrica 

**Otros nombres**: Variable numrica o categrica



**Ejemplos de uso**
- Como un doctor, me gustara saber si el nivel de colesterol de una persona correlaciona con su peso.

#### Cundo Utilizar
- Cuando quiero comprobar si existe algun tipo de correlacin entre dos variables (ver imagen de tipos de correlaciones).
- Cuando quiero observar la distribucin de dos variables al mismo tiempo.
- Cuando quiero hallar el tipo, y el nivel (o fuerza) de correlacin entre mis varibles (ver imagen de tipos de correlaciones).



#### Consejos
- No es necesario que el Eje Y empiece en el valor del "cero". En realidad, ambos ejes deberan empezar donde empieza el rango de sus datos.
- No utlices colores en los puntos innecesariamente.
- Puedo introducir una tercera o cuarta variable al grfico con las siguientes tcnicas:
  - Diferentes colores de cada punto para introducir una variable categrica.
  - Diferentes formas en los puntos para introducir una variable categrica.
  - Convertirlo en un grfico de burbujas para introducir una variable numrica.
  - Jugar con la intensidad del color de cada punto para introducir una variable numrica.
- Si tengo muy pocos punto en mi dataset, el resultado del grfico no ser muy til.
- Si hay una cantidad muy grande de puntos, el grfico se vuleve ilegible. Con apariencia de manchas.
  - Podemos coger una muestra, o jugar con la opacidad de los puntos.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/pinguinos.csv')
df
```

```python
# Eliminamos las filas nulas
df = df.dropna()
df
```

### Pandas

```python
df.plot.scatter(
  x='flipper_length_mm',
  y='culmen_length_mm',
)

plt.title('Longitud de las Aletas vs Longitud del Pico de los Pinguinos')
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
df.plot.scatter(
  x='flipper_length_mm',
  y='culmen_length_mm',
  c= 'body_mass_g' # permite escoger otra columna del dataset y cambiarle el color
)

plt.title('Longitud de las Aletas vs Longitud del Pico de los Pinguinos')
```

### Plotly

```python
import plotly.express as px
```

```python
fig = px.scatter(
  df,
  x='flipper_length_mm',
  y='culmen_length_mm',
  color='species' # para difrenciar por variable categorica, en este caso la especie
)

fig.show()
```

#### Grfico de Dispersin Categrico
En seaborn hay un tipo de grfico especial que ayuda cuando quiero hacer un Scatter plot utilizando variables categricas como los ejes. Me es til si quiero determinar cuando una variable categrica puede estar relacionada a cambios en una variable numrica. Uno de los dos ejes tiene que ser una varible categrica y el otro una variable numrica.

**Eje X**: Variable numrica o categrica.
**Eje Y**: Variable numrica o categrica.
**Otros Nombres**: Categora plot.
