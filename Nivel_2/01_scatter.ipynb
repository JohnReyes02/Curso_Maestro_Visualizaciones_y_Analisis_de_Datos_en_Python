## Gráfico de Dispersión (Scatter Plot)
Gráfico que muestra los valores de dos variables de un mismo registro en nuestro dataset al mismo tiempo. En cada Eje del gráfico se coloca una variable. Se dibuja en el canvas un punto, en la coordenada correspondiente al valor de las variables por cada observación de nuestra dataset. Su objetivo es detectar si existe o no alguna correlación entre las variables en los ejes. Estos ejes usualmente son variables numéricas pero también se pueden utilizar variables categóricas ordinales.

**Eje X**: Variable numérica o categórica 

**Eje Y**: Variable numérica o categórica 

**Otros nombres**: Variable numérica o categórica



**Ejemplos de uso**
- Como un doctor, me gustaría saber si el nivel de colesterol de una persona correlaciona con su peso.

#### Cuándo Utilizar
- Cuando quiero comprobar si existe algun tipo de correlación entre dos variables (ver imagen de tipos de correlaciones).
- Cuando quiero observar la distribución de dos variables al mismo tiempo.
- Cuando quiero hallar el tipo, y el nivel (o fuerza) de correlación entre mis varibles (ver imagen de tipos de correlaciones).



#### Consejos
- No es necesario que el Eje Y empiece en el valor del "cero". En realidad, ambos ejes deberían empezar donde empieza el rango de sus datos.
- No utlices colores en los puntos innecesariamente.
- Puedo introducir una tercera o cuarta variable al gráfico con las siguientes técnicas:
  - Diferentes colores de cada punto para introducir una variable categórica.
  - Diferentes formas en los puntos para introducir una variable categórica.
  - Convertirlo en un gráfico de burbujas para introducir una variable numérica.
  - Jugar con la intensidad del color de cada punto para introducir una variable numérica.
- Si tengo muy pocos punto en mi dataset, el resultado del gráfico no será muy útil.
- Si hay una cantidad muy grande de puntos, el gráfico se vuleve ilegible. Con apariencia de manchas.
  - Podemos coger una muestra, o jugar con la opacidad de los puntos.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/pinguinos.csv')
df
```

```python
# Eliminamos las filas nulas
df = df.dropna()
df
```

### Pandas

```python
df.plot.scatter(
  x='flipper_length_mm',
  y='culmen_length_mm',
)

plt.title('Longitud de las Aletas vs Longitud del Pico de los Pinguinos')
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
df.plot.scatter(
  x='flipper_length_mm',
  y='culmen_length_mm',
  c= 'body_mass_g' # permite escoger otra columna del dataset y cambiarle el color
)

plt.title('Longitud de las Aletas vs Longitud del Pico de los Pinguinos')
```

### Plotly

```python
import plotly.express as px
```

```python
fig = px.scatter(
  df,
  x='flipper_length_mm',
  y='culmen_length_mm',
  color='species' # para difrenciar por variable categorica, en este caso la especie
)

fig.show()
```

#### Gráfico de Dispersión Categórico
En seaborn hay un tipo de gráfico especial que ayuda cuando quiero hacer un Scatter plot utilizando variables categóricas como los ejes. Me es útil si quiero determinar cuando una variable categórica puede estar relacionada a cambios en una variable numérica. Uno de los dos ejes tiene que ser una varible categórica y el otro una variable numérica.

**Eje X**: Variable numérica o categórica.
**Eje Y**: Variable numérica o categórica.
**Otros Nombres**: Categoría plot.
