## Joinplot
Es un gráfico compuesto por dos tipos de gráficos diferentes. Es un Scatter Plot que al mismo tiempo muestra histogramas (o KDE) con la distribución de cada variable. Para este gráfico, ambas variables deben ser numéricas.

**Eje X**: Variable numérica.

**Eje Y**: Variable numérica.



### En esta clase
#### Datos
Información sobre especies de pinguinos

#### Objetivos 
1. Encontrar alguna relación entre las características de los pinguinos y al mismo tiempo ver la distribución de las variables

### Referencias
- https://github.com/lkuffo/data-viz/blob/main/NIVEL2/3_jointplot.ipynb
- https://seaborn.pydata.org/generated/seaborn.jointplot.html

```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid")
```

```python
df = pd.read_csv('../data/pinguinos.csv')
df = df.dropna()
df
```

```python
sns.jointplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    hue='species' # Se codifica una variable categorica de diferente color
)
```

```python
# gráfico de densidades con la regresión lineal
sns.jointplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    kind='reg'
    # hue='species' # Se codifica una variable categorica de diferente color
)
```
