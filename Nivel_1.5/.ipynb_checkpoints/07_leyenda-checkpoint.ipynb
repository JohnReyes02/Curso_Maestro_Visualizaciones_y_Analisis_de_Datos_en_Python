## Leyenda
La leyenda es una componente auxiliar de cualquier visualización que ayuda al lector a facilitar la legibilidad y comprensión de la misma. En ésta se pueden mostrar:
- Colores/formas/anchos de puntos o líneas que codifican variables categóricas.
- Colores/anchos de puntos que codifican rangos de variables numéricas.
- Rango de valores que codifica una intensidad de color de un punto a una variable numérica.



### Referencias
- https://github.com/lkuffo/data-viz/blob/main/NIVEL1.5/7_leyenda.ipynb

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/tiendas_procesado.csv', index_col='fecha')
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use(
    'ggplot'
)

ax = df.plot.line(
    figsize=(10, 4),
)

ax.set_title(
    'Ventas de 3 tipos de productos en el año 2015',
    fontdict={
        'size': 18,
        # 'name': 'Helvetica',
    }
)

ax.set_xlabel(
    'Período',
    fontdict={
        # 'name': 'Helvetica',
        'color': 'black',
        'size': 12
    }
)

ax.set_ylabel(
    'Ventas (en millones de dólares)',
    fontdict={
    #    'name': 'Helvetic',
        'color': 'black',
        'size': 12
    }
)
    
ax.set_xticks(np.arange(0, len(df)))
ax.set_xticklabels(
    df.index.values,
    rotation=45,
    fontdict={
      #  'name': 'Helvetic',
        'color': 'grey',
        'size': 10
    }
)

# Añadir anotaciones
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.annotate.html
ax.annotate(
    'Noviembre OJO!',
    xy = (10, 1250000),
    xytext = (10, 1000000),
    color='grey',
    horizontalalignment='center',
    arrowprops={
        'width': 1,
        'color': 'black',
        'headwidth': 5,
        'headlength': 5         
    }
)

# Configuración de Leyenda
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
ax.legend(
    title='Tipo de Producto', # El titulo
    labels=['Lacteos', 'Carnes', 'Comida de mar'], # Se puede definir el texto de la leyenda
    fontsize=10, # tamanio de las letras de la leyenda
    loc='upper left', # Punto de referencia de la leyenda con respecto a las esquinas
    bbox_to_anchor=(0.25, -0.3), # En que posicion del eje X y Y se pone la referencia
    labelcolor=['red', 'blue', 'purple'], # Cambia el color de la letra en la leyenda
    title_fontproperties={ # Propiedades del titulo de la leyenda
        'size': 8
    },
    alignment='center', # Alineacion de la leyenda
    ncol=3 # Para mostrar la forma en que se meustran la leyenda
)
```

### Seaborn

```python
import seaborn as sns
```

```python
plt.figure(
    figsize=(8, 4)
)

plt.title('Ventas de 3 tipos de productos en el año 2015')
plt.xlabel('Período')
plt.ylabel('Ventas (en millones de dólares')

sns.set_style('ticks')
sns.set_theme('paper')

ax = sns.lineplot(
    data=df
)

# Se realiza similar a Matplotlib usando fontdict

# ax.set_title(
#     'TITLE',
#     fontdict={
#         'color': 'blue',
#         'weight': 'bold',
#         'size': 20,
#         'family': 'serif'
        
#     }
# )

# ax.set_xlabel(
#     'XLABEL',
#     fontdict={
#         'color': 'green',
#         'weight': 'bold',
#         'size': 20,
#         'family': 'serif'
        
#     }
# )

# Añadir anotaciones
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.annotate.html
# ax.annotate(
#     'Noviembre OJO!',
#     xy = (10, 1250000),
#     xytext = (10, 1000000),
#     color='grey',
#     horizontalalignment='center',
#     arrowprops={
#         'width': 1,
#         'color': 'black',
#         'headwidth': 5,
#         'headlength': 5         
#     }
# )

# Tambien se puede hacer una linea horizontal que recorra todo el eje

ax.axhline(
    1300000,
    linestyle='dashed',
    linewidth=0.7,
    alpha=0.5,
    color='red'
)

# ax.axvline(
#     11,
#     linestyle='dashed',
#     linewidth=0.7,
#     alpha=0.5,
#     color='red'
# )

ax.annotate(
    'Umbral',
    xy = (11, 1250000),
    horizontalalignment='center',
    color='red',
    fontstyle='italic',
    alpha=0.8,
    size=8    
)

ax.legend(
    title='Tipo de Producto', # El titulo
    labels=['Lacteos', 'Carnes', 'Comida de mar'], # Se puede definir el texto de la leyenda
    fontsize=10, # tamanio de las letras de la leyenda
    loc='upper left', # Punto de referencia de la leyenda con respecto a las esquinas
    bbox_to_anchor=(0.22, -0.15), # En que posicion del eje X y Y se pone la referencia
    labelcolor=['red', 'blue', 'purple'], # Cambia el color de la letra en la leyenda
    title_fontproperties={ # Propiedades del titulo de la leyenda
        'size': 8
    },
    alignment='center', # Alineacion de la leyenda
    ncol=3 # Para mostrar la forma en que se meustran la leyenda
)

# Para no mostrar la leyenda
ax.legend().set_visible(False)
```

### Plotly

```python
import plotly.express as px
```

```python
fig = px.line(
    df,
    width=800,
    height=500,
    template='ggplot2',
    title='Ventas de 3 tipos de producto por año'    
)

# Para cambios globales
fig.update_layout(
    xaxis_title='Período',
    yaxis_title='Ventas en millones de USD',
    font_color='blue',
    font_family='Courier New',
    legend_font_color='black',
    legend_title_font_color='green',
    title_font_color='red',
    title_font_family='Times New Roman',
)

# Para cambios en los ejes
fig.update_xaxes(
    tickangle=45,
    tickvals=df.index.values,
    tickmode='array',
    # Para el titulo de eje
    title_font_family='Arial',
    title_font_color='Black',
    title_font_size=26,
    # Para los ticks del eje
    tickfont={
        'family': 'Rockwell',
        'color': 'brown',
        'size': 10
    }
)

fig.update_yaxes(
    dtick=250000
)

fig.add_annotation(
    x='2015-11',
    y=1300000,
    text='Noviembre!',
    opacity=0.5,
    font={
        'size': 12,
        'color': 'black'
    },
    showarrow=True,
    arrowsize=1,
    arrowwidth=2,
    arrowhead=2 # es ID del 0 al 8, que indica el tipo de la flecha
)

fig.add_hline(
    y=1300000,
    annotation_text = 'Limite',
    annotation_position='top left',
    annotation_font_color='red',
    annotation_font_size=10,
    line_dash='dot',
    opacity=0.8,
    line_width=1,
    line_color='red'
)

# Configuracion de la leyenda
fig.update_layout(
    legend_title_text='Leyenda',
    legend={
        'y': 0.5,
        'x': 1.05,
        'font':{
            'family': 'Curier',
            'size': 12,
            'color': 'black'
        },
        'bgcolor': 'LightGrey',
        'bordercolor': 'Black',
        'borderwidth': 2
    }
)

fig.show()
```
