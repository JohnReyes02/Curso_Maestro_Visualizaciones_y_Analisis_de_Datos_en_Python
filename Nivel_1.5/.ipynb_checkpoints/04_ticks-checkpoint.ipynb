## Ticks
Se denominan ticks al conjunto de etiquetas en el Eje X y Y.
- Su utilidad es poder codificar los valores o punto del gráfico a valores específicos en el Eje X y Y.
- Un gráfico sin ticks es ilegible.

  

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/tiendas_procesado.csv', index_col='fecha')
df
```

### Pandas

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
posiciones_ticks = np.arange(0, len(df))
posiciones_ticks
df.index.values
```

```python
plt.style.use(
    'ggplot'
)

ax = df.plot.line(
    figsize=(10, 5),
    title='Ventas de 3 tipos de productos en el año 2015'
)

# Primer paso: en que posición quiero las etiquetas
ax.set_xticks(
    posiciones_ticks
)

# Segundo paso: Cuáles son las etiquetas
ax.set_xticklabels(
    df.index.values,
    rotation=45
)

# Para nombrar el eje X
ax.set_xlabel(
    'Período',
)

# Para nombrar el eje Y
ax.set_ylabel(
    'Ventas (en millones de dólares)',
)
```

```python
# Si quiero etiquetas salteadas

plt.style.use(
    'ggplot'
)

ax = df.plot.line(
    figsize=(10, 5),
    title='Ventas de 3 tipos de productos en el año 2015'
)

# Primer paso: en que posición quiero las etiquetas
ax.set_xticks(
    posiciones_ticks[::2]
)

# Segundo paso: Cuáles son las etiquetas
ax.set_xticklabels(
    df.index.values[::2],
    rotation=45
)

# Para nombrar el eje X
ax.set_xlabel(
    'Período',
)

# Para nombrar el eje Y
ax.set_ylabel(
    'Ventas (en millones de dólares)',
)
```

```python
# Si quiero etiquetas salteadas en el eje Y

plt.style.use(
    'ggplot'
)

ax = df.plot.line(
    figsize=(10, 5),
    title='Ventas de 3 tipos de productos en el año 2015'
)

ax.set_yticks(
    [500000, 1000000]
)

ax.set_yticklabels(
    ['MEDIO MILLON', 'UN MILLON']
)

# Para nombrar el eje X
ax.set_xlabel(
    'Período',
)

# Para nombrar el eje Y
ax.set_ylabel(
    'Ventas (en millones de dólares)',
)
```

### Seaborn

```python
import seaborn as sns
```

```python
plt.figure(
    figsize=(8, 4)
)

plt.title('Ventas de 3 tipos de productos en el año 2015')
plt.xlabel('Periíodo')
plt.ylabel('Ventas (en millones de dólares)')

sns.set_style('ticks')
sns.set_theme('paper')

ax = sns.lineplot(
    data=df
)

ax.set_xticks([0, 1])
ax.set_xticklabels(['ENERO', 'FEBRERO'], rotation=90)
```

### Plotly

```python
import plotly.express as px
```

```python
fig = px.line(
    df,
    width=800,
    height=500,
    template='ggplot2',
    title='Ventas de 3 tipos de porductos por año'
)

fig.update_layout(
    xaxis_title='Período',
    yaxis_title='Ventas en millones de USD'
)

fig.update_xaxes(
    tickangle=45,
    tickvals=df.index.values,
    tickmode='array'
)

fig.update_yaxes(
    # nticks=8
    # tickvals=[500000, 1000000]
    dtick=250000
)

fig.show()
```
