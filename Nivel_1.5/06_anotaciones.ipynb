## Anotaciones
A veces desearemos escribir notas en el grfico o resltar algn punto de inters con texto.



Tambin es til utilizar `marcas` como lneas verticales u horizontales para resaltar lmites o umbrales.

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/tiendas_procesado.csv', index_col='fecha')
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use(
    'ggplot'
)

ax = df.plot.line(
    figsize=(10, 4),
)

ax.set_title(
    'Ventas de 3 tipos de productos en el ao 2015',
    fontdict={
        'size': 18,
        # 'name': 'Helvetica',
    }
)

ax.set_xlabel(
    'Perodo',
    fontdict={
        # 'name': 'Helvetica',
        'color': 'black',
        'size': 12
    }
)

ax.set_ylabel(
    'Ventas (en millones de dlares)',
    fontdict={
    #    'name': 'Helvetic',
        'color': 'black',
        'size': 12
    }
)
    
ax.set_xticks(np.arange(0, len(df)))
ax.set_xticklabels(
    df.index.values,
    rotation=45,
    fontdict={
      #  'name': 'Helvetic',
        'color': 'grey',
        'size': 10
    }
)

# Aadir anotaciones
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.annotate.html
ax.annotate(
    'Noviembre OJO!',
    xy = (10, 1250000),
    xytext = (10, 1000000),
    color='grey',
    horizontalalignment='center',
    arrowprops={
        'width': 1,
        'color': 'black',
        'headwidth': 5,
        'headlength': 5         
    }
)
```

### Plotly

```python
import plotly.express as px
```

```python
fig = px.line(
    df,
    width=800,
    height=500,
    template='ggplot2',
    title='Ventas de 3 tipos de producto por ao'    
)

# Para cambios globales
fig.update_layout(
    xaxis_title='Perodo',
    yaxis_title='Ventas en millones de USD',
    font_color='blue',
    font_family='Courier New',
    legend_font_color='black',
    legend_title_font_color='green',
    title_font_color='red',
    title_font_family='Times New Roman',
)

# Para cambios en los ejes
fig.update_xaxes(
    tickangle=45,
    tickvals=df.index.values,
    tickmode='array',
    # Para el titulo de eje
    title_font_family='Arial',
    title_font_color='Black',
    title_font_size=26,
    # Para los ticks del eje
    tickfont={
        'family': 'Rockwell',
        'color': 'brown',
        'size': 10
    }
)

fig.update_yaxes(
    dtick=250000
)

fig.add_annotation(
    x='2015-11',
    y=1300000,
    text='Noviembre!',
    opacity=0.5,
    font={
        'size': 12,
        'color': 'black'
    },
    showarrow=True,
    arrowsize=1,
    arrowwidth=2,
    arrowhead=2 # es ID del 0 al 8, que indica el tipo de la flecha
)

fig.add_hline(
    y=1300000,
    annotation_text = 'Limite',
    annotation_position='top left',
    annotation_font_color='red',
    annotation_font_size=10,
    line_dash='dot',
    opacity=0.8,
    line_width=1,
    line_color='red'
)

fig.show()
```
