## Subgráficos
Es posible dibujar varios gráficos en un mismo canvas. Esto se logra a través de los subgráficos (o subplots).

- Cada subgráfico se encontrará en una variable. Con esta yo puedo acceder a los métodos de graficación y personalización de solamente ese subgráfico.
- Los subgráficos se acomodan en una cuadrícula de dimensiones predefinidas.
- En este momento vamos a diferenciar dos tipos de métodos.
  **Métodos de figura**:
  Alternan propiedades de todo el canvas.
  **Métodos de Axes**:
  Alternan propiedades de solamente un subgráfico.

```python
# Desempacando los subgraficos en variables individuales
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, constrained_layout=True)

fig.suptitle('Venta de 3 tipos de productos')
fig.supylabel('Venta en millones de dolares')
fig.supxlabel('Periodo')

fig.set_size_inches(6, 4)

plt.style.use(
    'ggplot'
)

df.plot.line(
    y='DAIRY',
    ax=ax1,
    color='blue'
)

df.plot.line(
    y='MEATS',
    ax=ax2,
    color='red'
)

df.plot.line(
    y='SEAFOOD',
    ax=ax3,
    color='green'
)

ax1.set_xticklabels([]) # Quita los ticks del eje X de la primer grafica
ax2.set_xticklabels([]) # Quita los ticks del eje X de la segundo grafica

ax1.set_xlabel('')
ax2.set_xlabel('')
ax3.set_xlabel('')

ax1.set_title(
    'Lacteos',
    fontdict={
        'size': 8,
        'color': 'blue'
    }
)
ax2.set_title(
    'Carnes', 
    fontdict={
        'size': 8,
        'color': 'red'
    }
)
ax3.set_title(
    'Comida de Mar', 
    fontdict={
        'size': 8,
        'color': 'green'
    }
)

ax1.legend().set_visible(False)
ax2.legend().set_visible(False)
ax3.legend().set_visible(False)



# ax.set_title(
#     'Ventas de 3 tipos de productos en el año 2015',
#     fontdict={
#         'size': 18,
#         # 'name': 'Helvetica',
#     }
# )

# ax.set_xlabel(
#     'Período',
#     fontdict={
#         # 'name': 'Helvetica',
#         'color': 'black',
#         'size': 12
#     }
# )

# ax.set_ylabel(
#     'Ventas (en millones de dólares)',
#     fontdict={
#     #    'name': 'Helvetic',
#         'color': 'black',
#         'size': 12
#     }
# )
    
# ax.set_xticks(np.arange(0, len(df)))
# ax.set_xticklabels(
#     df.index.values,
#     rotation=45,
#     fontdict={
#       #  'name': 'Helvetic',
#         'color': 'grey',
#         'size': 10
#     }
# )

# # Añadir anotaciones
# # https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.annotate.html
# ax.annotate(
#     'Noviembre OJO!',
#     xy = (10, 1250000),
#     xytext = (10, 1000000),
#     color='grey',
#     horizontalalignment='center',
#     arrowprops={
#         'width': 1,
#         'color': 'black',
#         'headwidth': 5,
#         'headlength': 5         
#     }
# )

# # Configuración de Leyenda
# # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
# ax.legend(
#     title='Tipo de Producto', # El titulo
#     labels=['Lacteos', 'Carnes', 'Comida de mar'], # Se puede definir el texto de la leyenda
#     fontsize=10, # tamanio de las letras de la leyenda
#     loc='upper left', # Punto de referencia de la leyenda con respecto a las esquinas
#     bbox_to_anchor=(0.25, -0.3), # En que posicion del eje X y Y se pone la referencia
#     labelcolor=['red', 'blue', 'purple'], # Cambia el color de la letra en la leyenda
#     title_fontproperties={ # Propiedades del titulo de la leyenda
#         'size': 8
#     },
#     alignment='center', # Alineacion de la leyenda
#     ncol=3 # Para mostrar la forma en que se meustran la leyenda
# )
```

### Seaborn

```python
import seaborn as sns
```

```python
# Desempacando los subgraficos en variables individuales
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, constrained_layout=True)

fig.suptitle('Ventas de 3 tipos de productos en el año 2015')
fig.supxlabel('Período')
fig.supylabel('Ventas (en millones de dólares')

sns.set_style('ticks')
sns.set_theme('paper')

sns.lineplot(
    data=df['DAIRY'],
    ax=ax1, 
    color='blue'
)
sns.lineplot(
    data=df['MEATS'],
    ax=ax2,
    color='red'
)
sns.lineplot(
    data=df['SEAFOOD'],
    ax=ax3,
    color='green'
)

ax1.set_xticklabels([]) # Quita los ticks del eje X de la primer grafica
ax2.set_xticklabels([]) # Quita los ticks del eje X de la segundo grafica

ax1.set_xlabel('')
ax2.set_xlabel('')
ax3.set_xlabel('')

ax1.set_title(
    'Lacteos',
    fontdict={
        'color': 'blue',
        'size': 8
    }
)

ax2.set_title(
    'Carnes',
    fontdict={
        'color': 'red',
        'size': 8
    }
)

ax3.set_title(
    'Comida de Mar',
    fontdict={
        'color': 'green',
        'size': 8
    }
)


# Se realiza similar a Matplotlib usando fontdict

# ax.set_title(
#     'TITLE',
#     fontdict={
#         'color': 'blue',
#         'weight': 'bold',
#         'size': 20,
#         'family': 'serif'
        
#     }
# )

# ax.set_xlabel(
#     'XLABEL',
#     fontdict={
#         'color': 'green',
#         'weight': 'bold',
#         'size': 20,
#         'family': 'serif'
        
#     }
# )

# Añadir anotaciones
# https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.annotate.html
# ax.annotate(
#     'Noviembre OJO!',
#     xy = (10, 1250000),
#     xytext = (10, 1000000),
#     color='grey',
#     horizontalalignment='center',
#     arrowprops={
#         'width': 1,
#         'color': 'black',
#         'headwidth': 5,
#         'headlength': 5         
#     }
# )

# Tambien se puede hacer una linea horizontal que recorra todo el eje

ax1.axhline(
    1300000,
    linestyle='dashed',
    linewidth=0.7,
    alpha=0.5,
    color='red'
)

# ax.axvline(
#     11,
#     linestyle='dashed',
#     linewidth=0.7,
#     alpha=0.5,
#     color='red'
# )

ax1.annotate(
    'Umbral',
    xy = (11, 1250000),
    horizontalalignment='center',
    color='red',
    fontstyle='italic',
    alpha=0.8,
    size=8    
)

ax1.legend(
    title='Tipo de Producto', # El titulo
    labels=['Lacteos', 'Carnes', 'Comida de mar'], # Se puede definir el texto de la leyenda
    fontsize=10, # tamanio de las letras de la leyenda
    loc='upper left', # Punto de referencia de la leyenda con respecto a las esquinas
    bbox_to_anchor=(0.22, -0.15), # En que posicion del eje X y Y se pone la referencia
    labelcolor=['red', 'blue', 'purple'], # Cambia el color de la letra en la leyenda
    title_fontproperties={ # Propiedades del titulo de la leyenda
        'size': 8
    },
    alignment='center', # Alineacion de la leyenda
    ncol=3 # Para mostrar la forma en que se meustran la leyenda
)

# Para no mostrar la leyenda
ax1.legend().set_visible(False)
```

### Plotly

```python
import plotly.express as px # No soporta subgraficos
```

```python
fig = px.line(
    df,
    width=800,
    height=500,
    template='ggplot2',
    title='Ventas de 3 tipos de producto por año'    
)

# Para cambios globales
fig.update_layout(
    xaxis_title='Período',
    yaxis_title='Ventas en millones de USD',
    font_color='blue',
    font_family='Courier New',
    legend_font_color='black',
    legend_title_font_color='green',
    title_font_color='red',
    title_font_family='Times New Roman',
)

# Para cambios en los ejes
fig.update_xaxes(
    tickangle=45,
    tickvals=df.index.values,
    tickmode='array',
    # Para el titulo de eje
    title_font_family='Arial',
    title_font_color='Black',
    title_font_size=26,
    # Para los ticks del eje
    tickfont={
        'family': 'Rockwell',
        'color': 'brown',
        'size': 10
    }
)

fig.update_yaxes(
    dtick=250000
)

fig.add_annotation(
    x='2015-11',
    y=1300000,
    text='Noviembre!',
    opacity=0.5,
    font={
        'size': 12,
        'color': 'black'
    },
    showarrow=True,
    arrowsize=1,
    arrowwidth=2,
    arrowhead=2 # es ID del 0 al 8, que indica el tipo de la flecha
)

fig.add_hline(
    y=1300000,
    annotation_text = 'Limite',
    annotation_position='top left',
    annotation_font_color='red',
    annotation_font_size=10,
    line_dash='dot',
    opacity=0.8,
    line_width=1,
    line_color='red'
)

# Configuracion de la leyenda
fig.update_layout(
    legend_title_text='Leyenda',
    legend={
        'y': 0.5,
        'x': 1.05,
        'font':{
            'family': 'Curier',
            'size': 12,
            'color': 'black'
        },
        'bgcolor': 'LightGrey',
        'bordercolor': 'Black',
        'borderwidth': 2
    }
)

fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
fig = make_subplots(rows=3, cols=1)

fig.append_trace(
    go.Scatter(
        x=df.index.values,
        y=df['DAIRY'],
        name='LACTEOS'
    ), row=1, col=1
    
)

fig.append_trace(
    go.Scatter(
        x=df.index.values,
        y=df['MEATS'],
        name='CARNES'
    ), row=2, col=1
    
)
fig.append_trace(
    go.Scatter(
        x=df.index.values,
        y=df['SEAFOOD'],
        name='COMIDA DE MAR'
    ), row=3, col=1
    
)

fig.update_layout(
    height=600,
    width=700,
    title_text='Venta de 3 tipos de productos'
)

fig.show()
```
