## Grid (Cuadrcula)
Lneas opacas que sirven como ayuda al lector para codificar los valores o puntos del grfico a los valores en los ejes.
- Es un elemento opcional de una visualizacin. Pero usualmente es recomendable utilizarlo.
- Si utilizamos una plantilla, esta va a configurar automticamente los estilos de la cuadrcula.
- Podemos configurar el estilo de las lneas del Grid, la opacidad, colores y en qu eje aparecen.



```python
import pandas as pd
```

```python
df = pd.read_csv('../data/tiendas_procesado.csv', index_col='fecha')
df
```

#### Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
plt.style.use(
    'bmh'
)

ax = df.plot.line(
    figsize=(10, 5)
)

# Cambiamos la cuadricula del eje x. Por ejemplo azul
ax.xaxis.grid(
   color='blue',
   alpha=0.2, # Porcentaje de opacidad
   linewidth=0.5,
)

#Cambiamos la cuadricula del eje y
ax.yaxis.grid(
    color='#28bf50',
    linestyle='-.'
)
```

#### Seaborn

```python
import seaborn as sns
```

```python
plt.figure(
    figsize=(8, 4)
)

sns.set_style('ticks')
sns.set_theme('paper')

plt.grid(
    visible=False,
    axis='x'    
)
plt.grid(
    axis='y',
    color='red',
    alpha=0.2
)

ax = sns.lineplot(
    data=df
)
```

#### Plotly

```python
import plotly.express as px
```

```python
fig = px.line(
    df,
    width=800,
    height=500,
    template='plotly_dark'
)

fig.update_xaxes(showgrid=False)
fig.update_yaxes(
    gridwidth=1,
    gridcolor='white',
    griddash='solid',
    minor_griddash='dot'
)
```
