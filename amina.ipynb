### Carga de Datos

```python
import pandas as pd
```

Se cambian todos los datos de objetos a formato numerico

```python
df = df.apply(pd.to_numeric, errors='coerce')
df
```

```python
df.info()
```

```python
# Filtramos las columnas que no queremos analizar y eliminamos los datos nulos
df_filtrado = df.drop(columns = ['109-FIC-1905.MEAS', '109-FIC-1906.MEAS'])
df_filtrado = df_filtrado.dropna()
df_filtrado
```

```python
# Omitimos los datos donde el flujo de amina es menor a 80 gpm
df_filtrado_on = df_filtrado[(df_filtrado['109-FIC-3707.MEAS'] > 100) & (df_filtrado['109-FIC-3707.MEAS'] < 120)]
df_filtrado_on
```

```python
matriz = df_filtrado_on.corr()
matriz
```

```python
import seaborn as sns
```

```python
sns.heatmap(
    matriz,
    cmap='vlag',
    annot=True
)
```

```python
sns.heatmap(
    matriz,
    cmap='RdBu',
    square=True,
    annot=True,
    vmin=-1,
    vmax=1
)
```
