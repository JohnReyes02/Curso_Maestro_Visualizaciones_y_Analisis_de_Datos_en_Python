## Transformacion de datos

```python
# Calcular una nueva columna a partir de una columna

import random
def calcular_ganancias(retweets: int) -> int:
  """Fucion que calcula el nuevo valor de ganancias
     a partir de multiplicar el valor de retweets con un 
     numero aleatorio entre 3 y 5   

  Args:
      retweets (int): Indica el numero que de retweet que viene
      de la columna del df

  Returns:
      int: Retorna la variable ganancia para la nueva columna
  """
  ganancia = retweets * random.randint(3, 5)
  return ganancia


df['ganancias'] = df['retweets'].apply(calcular_ganancias)

df.head()
```

```python
# Calcular nueva columna a partir de varias columnas

def popularidad(fila: int) -> float:
  """Calcula la popularidad dividiendo las personas que sigo'
    entre las personas que me siguen

  Args:
      fila (int): Lee la fila que se llame desde el dataframe

  Returns:
      float: Se obtiene calculando followees / followers
  """  
  resultado = fila['followees'] / fila['followers']
  return resultado

df['popularidad'] = df.apply(popularidad, axis=1)
df.head()
```

## Agrupacion de datos
Por un valor de una columna, por ejemplo con datos categoricos

Ej: Calcular el promedio de favoritos en los tweets por cada pais. Para esto se utiliza la funcion `groupby()`.

```python
df[
    ['country', 'favorites', 'retweets', 'mentions', 'followers', 'followees', 'ganancias', 'popularidad']
  ].groupby('country').mean()
```

```python
# Se puede especificar el tipo de agrupacion para cada columna
df.groupby('country').agg({
  'followers': 'sum',
  'mentions': 'mean',
  'retweets': 'max'
})
```

Filtrado de agrupacion de datos

```python
grouped = df.groupby('country').agg({
  'followers': 'sum',
  'mentions': 'mean',
  'retweets': 'max'
})

grouped[grouped['followers'] > 5000]
```
