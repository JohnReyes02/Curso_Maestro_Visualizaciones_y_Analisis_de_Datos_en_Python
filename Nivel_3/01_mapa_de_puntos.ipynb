## Mapa de puntos
Grfico que muestra en un espacio geogrfico la distribucin de puntos de inters. Su objetivo es mostrar como se distribuyen mis entidades de inters (estaciones, casos mdicos, edificios, etc) a lo largo de un espacio geogrfico (mapa). Para poder graficarlo necesito la posicin geogrfica de cada punto.



### Qu necesito para hacer un Mapa de Puntos?
- Necesito un archivo en donde mis registros tengan coordenadas geogrficas en cualquier formato.

### Consejos
- Puedo introducir una o ms variables al grfico con las siguientes tcticas:
    - Diferentes formas en los "puntos" para introducir una variable categrica.
    - Convertir los puntos en burbujas de diferentes tamaos para introducir una variable numrica. Estos es un **Mapa de Burbujas**.
    - Jugar con la intensidad del color de cada punto para introducir una variable numrica.
- Si conecto los puntos con lneas por alguna relacin que estos tengan, o por algn movimiento, se convierte en un Mapa de Conexiones o Flujo.
- Ti tengo una alta densidad de puntos, y empiezo a ver solo manchas, puedo considerar utilizar un HeatMap Geogrfico, o hacer agrupacin de puntos.
