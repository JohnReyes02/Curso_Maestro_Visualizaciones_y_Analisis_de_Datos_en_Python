## Grfico de Barras Apiladas
Es un grfico de barras en donde apilamos las barras de distintas categoras en una misma categora codificada en el Eje X. Tiene la misma utilidad del grfico de barras pero podemos visualizar cambios de magnitud de diversas categoras al mismo tiempo.

**Eje X**: Variable Categrica

**Eje Y**: Variable Numrica

**Otros nombres**: Grfico de Columnas Apiladas, Stacked Bar Chart



#### Cuando Utilizar
- Cuando quiero analizar los cambios de magnitud de diversas categoras al mismo tiempo a lo largo de otra variable categrica.
- Cuando me interesa ver al mismo tiempo los cambios de la sumatoria total de la variable numrica de entre todas las categoras.

#### Consejos
- No apilar demasiadas categoras ya que el grfico se vuelve difcil de interpretar.
- Es recomendable utilizar un color diferente para diferenciar cada categora de las barras apiladas.
- En todas las barras guarda el mismo orden de apilado.
- _Luego tenemos los mismos consejos que en un grfico de barras comn..._
- Empieza el eje Y en el valor del "cero". As no vas a "disfrazar" cambios de magnitudes irrelevantes **(Ver primera imagen de abajo)**.
- Si las variables categricas del Eje X guardan un orden (Ejemplo: aos, meses, horas), se recomienda ordenar las columnas. Del mismo modo si quieres enfatizar que es un "Top" (en cuyo caso las barras se ordenarian segun el valor en el Eje Y).
- No hacer las barras "tridimensionales" o con formas extraas que pueden confundir al lector acerca  del valor en el eje Y **(Ver segunda imagen de abajo)**.
- Barras verticales o horizontales? No hay preferencia. Puede depender de cuntas categoras quiero mostrar o del tamao de mis etiquetas categricas.
- Se en ves de apilar las barras, las coloco en un costado una despus de la otra se convierte en un Grfico de Barras Mltiples (ver grfico).
