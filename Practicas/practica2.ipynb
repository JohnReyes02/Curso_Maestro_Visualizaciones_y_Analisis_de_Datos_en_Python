## Práctica 2: Barras Apiladas

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../data/datasets_top_4.csv', index_col='dataset')
df
```

#### Algunos tips
- utilizar plot.barh, parametro stacked=True
- La leyenda tiene forma de columna
- Hay un solo titulo para los dos graficos
- En el primer subgrafico las etiquetas estan a la izquierda del eje y
- En el segundo subgrafico las etiquetas estan a la derecha del eje y

### Mi solucion

```python
df1 = df[:len(df)//2]
df1
```

```python
df2 = df[len(df)//2:]
df2
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True)
fig.suptitle('Como se distibuyen las categorias dentro de los datasets?')
fig.supylabel('Dataset')

fig.set_size_inches(7, 4)

df1.plot(
    kind='barh',
    ax=ax1,
    stacked=True
)

df2.plot(
    kind='barh',
    ax=ax2,
    stacked=True
)

ax1.set_xlabel('Porcentaje')
ax2.set_xlabel('Porcentaje')
ax1.set_ylabel('')
ax2.set_ylabel('')

ax1.legend().set_visible(False)

ax2.legend(
    labels=['Categoria 1', 'Categoria 2', 'Categoria 2', 'Otros'], # Se puede definir el texto de la leyenda
    # fontsize=10, # tamanio de las letras de la leyenda
    loc='upper left', # Punto de referencia de la leyenda con respecto a las esquinas
    bbox_to_anchor=(-2.2, -0.2), # En que posicion del eje X y Y se pone la referencia
    # labelcolor=['red', 'blue', 'purple'], # Cambia el color de la letra en la leyenda
    # title_fontproperties={ # Propiedades del titulo de la leyenda
    #     'size': 8
    # },
    alignment='center', # Alineacion de la leyenda
    ncol=4 # Para mostrar la forma en que se meustran la leyenda
)
```

### La solucion presentada

```python
len(df)
```

```python
df1 = df.iloc[0:13, :] # division de DF primera parte
df2 = df.iloc[13:, :] # division del DF segunda parte
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

# Cambiamos el tamanio del canvas
fig.set_size_inches(7, 4)

# Titulo general del grafico
fig.suptitle('Como se distribuyen las categoria \ndentro de los datasets?')

df1.plot.barh(
    stacked=True, # Se apilan en una sola barra
    ax=ax1,
    width=0.4,
    color={ # Se define el color de cada columna
        'cat1': '#000086',
        'cat2': '#5396f4',
        'cat3': '#a1dbe8',
        'otros': '#DADADA'
    }
)

df2.plot.barh(
    stacked=True,
    ax=ax2,
    width=0.4,
    color={
        'cat1': '#000086',
        'cat2': '#5396f4',
        'cat3': '#a1dbe8',
        'otros': '#DADADA'
    }
)

ax2.set_ylabel('')
ax1.set_ylabel('Dataset')

ax1.set_xlabel('Porcentaje')
ax2.set_xlabel('Porcentaje')

ax1.legend().set_visible(False)

ax2.legend(
    ncol=4,
    labels=['Categoria 1', 'Categoria 2', 'Categoria 3', 'Otros'],
    loc='upper left',
    bbox_to_anchor=(-1.2, -0.15)
)

ax2.yaxis.tick_right() # Pone los ticks a la derecha

ax1.text( # Para agregar un texto dentro del grafico
    75, -5,
    'Algunos datos han sido ignorados del analisis',
    fontsize=6,
    bbox={
        'facecolor': 'orange',
        'alpha': 0.5
    }
)
```
