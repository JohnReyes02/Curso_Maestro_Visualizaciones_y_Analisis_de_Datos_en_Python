## Práctica 1: Barras Múltiples
Con un dataset  de rendimiento de algoritmos intentaremos comparar el rendimiento de ambos algoritmos en varias pruebas con un gráfico de barras múltiples.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/comparasion_algoritmos.csv')
df
```

#### Algunos tips
- No se manejan subgraficos
- La Leyenda tiene etiquetas diferentes
- Realizarlas con Pandas
- Las barras multiples en el eje Y se puede pasar una lista de columnas para lograr multiples barras.

```python
# ax = df.plot.bar(
#     figsize=(12, 5),
#     y=['algoritmo1', 'algoritmo2'],
#     x='dataset'
# )
# otra forma
ax = df.plot(
    kind='bar',
    figsize=(12, 5),
    y=['algoritmo1', 'algoritmo2'],
    x='dataset'
)

ax.set_title(
    'Evaluación de Algoritmo 1 vs Algoritmo 2'
)

ax.set_xlabel(
    'Dataset'
)

ax.set_ylabel(
    'Rendimiento'
)

ax.grid(
    visible=True
)

ax.legend(
    labels=['Algortitmo 1', 'Algoritmo 2']
)

# Para enviar hacia atras la cuadricula
ax.set_axisbelow(True)
```
